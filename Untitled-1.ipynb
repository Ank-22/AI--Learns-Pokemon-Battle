```python
import pandas as pd 
import matplotlib.pyplot as plt 
```

```python
data = pd.read_csv('bots/pokemon-rl/a3c_rewards.csv')
```

```python
data.head()
```

```python
data.info()
```

```python
data_values = data['values']

# Calculate the moving average with a window size of 100
window_size = 100
smoothed_data = data_values.rolling(window=window_size).mean()

# Plot the original data and the smoothed data
plt.figure(figsize=(10, 6))
plt.plot(data_values, label="Original Data", alpha=0.5)
plt.plot(smoothed_data, label="Smoothed Data (Moving Average)", color='red', linewidth=2)

plt.title("A3C")
plt.xlabel("timesteps")
plt.ylabel("Rewards")
plt.legend()
plt.grid(True)
plt.show()
```
